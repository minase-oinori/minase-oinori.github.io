```python
import pandas as pd
import random
from datetime import datetime, timedelta

male_names = [
    "若山 智", "西野 悠史", "金丸 直人", "澤美 悠平", "武田 成和", 
    "平山 一夫", "本間 弦都", "杉本 功", "田熊 英嗣", "水野 英和"
]
female_names = [
    "若林 祐江", "西野 梨", "美華子", "澤美 嘉子", "和泉 祥子", 
    "切通 紗代子", "道井 理恵子", "久森 苑日", "黒柳 小友理", "田熊 美恵"
]

names_and_sexes = [(name, 'male') for name in male_names] + [(name, 'female') for name in female_names]
random.shuffle(names_and_sexes)
selected = names_and_sexes[:10]

# 年齢、欠損1つ、80歳1つ
ages = [random.randint(20, 40) for _ in range(8)]
ages.append(None)
ages.append(80)
random.shuffle(ages)

date_patterns = [
    "%Y-%m-%d",
    "%Y年%m月%d日",
    "%Y/%m/%d"
]

def random_date():
    start = datetime(2020, 1, 1)
    end = datetime(2025, 5, 31)
    delta = end - start
    r = random.randint(0, delta.days)
    date = start + timedelta(days=r)
    fmt = random.choice(date_patterns)
    return date.strftime(fmt)

# 2つ性別をランダムで欠損
sex_missing_indices = random.sample(range(10), 2)

rows = []
for i in range(10):
    name, sex = selected[i]
    sex_val = "" if i in sex_missing_indices else sex
    age_val = ages[i]
    date_val = random_date()
    row = {
        "name": name,
        "sex": sex_val,
        "age": age_val,
        "register_date": date_val
    }
    rows.append(row)

df = pd.DataFrame(rows)

# ★保存先指定d
save_path = r'C:\Users\hj7724\Downloads\member_list.csv'
df.to_csv(save_path, index=False, encoding='utf-8-sig')

print(f'CSVを保存しました: {save_path}')
print(df)
```
