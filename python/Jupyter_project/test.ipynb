```python
import pandas as pd
import random
import numpy as np
from datetime import datetime, timedelta

male_names = [
    "若山 智", "水野 英和", "金丸 直人", "澤美 悠平", "武田 成和", 
    "平山 一夫", "本間 弦都", "杉本 功", "田熊 英嗣", "水野 英和"
]
female_names = [
    "若林 祐江", "西野 梨", "美華子", "澤美 嘉子", "和泉 祥子", 
    "切通 紗代子", "道井 理恵子", "久森 苑日", "黒柳 小友理", "田熊 美恵"
]

names_and_sexes = [(name, 'male') for name in male_names] + [(name, 'female') for name in female_names]
random.shuffle(names_and_sexes)
selected = names_and_sexes  # 20人分

date_patterns = [
    "%Y-%m-%d",
    "%Y年%m月%d日",
    "%Y/%m/%d"
]

def random_date():
    start = datetime(2020, 1, 1)
    end = datetime(2025, 5, 31)
    delta = end - start
    r = random.randint(0, delta.days)
    date = start + timedelta(days=r)
    fmt = random.choice(date_patterns)
    return date.strftime(fmt)

# --- sex列特殊値 ---
SPECIAL_SEX = ["", np.nan, "-", "不明"]
n_sex_special = 8    # 特殊値8行
n_sex_normal = 12    # 正常値12行

# sex用のリスト作成
sex_values = [names_and_sexes[i][1] for i in range(n_sex_normal)]
sex_values += random.choices(SPECIAL_SEX, k=n_sex_special)
random.shuffle(sex_values)  # ランダムな配置

# --- age列 ---
SPECIAL_AGE = ["", np.nan, "-", "不明"]
n_age_special = 8
n_age_normal = 12

age_values = [random.randint(20, 40) for _ in range(n_age_normal)]
age_values += random.choices(SPECIAL_AGE, k=n_age_special-2)  # 残り6個くらいは欠損・文字
age_values.append(None)            # None(欠損)も念のため加える
age_values.append(9999)            # 1つは9999
random.shuffle(age_values)

# points: 0~1000を18個、10000~20000を2個
points_list = [random.randint(0, 1000) for _ in range(18)] + \
              [random.randint(10000, 20000) for _ in range(2)]
random.shuffle(points_list)

# sale列を作成
sale_list = [random.randint(1, 100000) for _ in range(17)]    # 1～100,000 の17個
sale_list += [random.randint(2000000, 5000000) for _ in range(3)]  # 2,000,000～5,000,000の3個
random.shuffle(sale_list)

rows = []
for i in range(20):
    name = selected[i][0]
    sex_val = sex_values[i]
    age_val = age_values[i]
    date_val = random_date()
    point_val = points_list[i]
    sale_val = sale_list[i]
    row = {
        "name": name,
        "sex": sex_val,
        "age": age_val,
        "register_date": date_val,
        "points": point_val,
        "sale": sale_val      # 追加
    }
    rows.append(row)

df = pd.DataFrame(rows)

# 保存先
save_path = r'C:\Users\hj7724\VScode\python\pj\data\member_list_20250605.csv'
df.to_csv(save_path, index=False, encoding='utf-8-sig')

print(f'CSVを保存しました: {save_path}')
print(df)
```
